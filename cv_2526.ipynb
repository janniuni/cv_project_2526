## Data

```python
DATA_DIR = "./data"
os.makedirs(DATA_DIR, exist_ok=True)

ZENODO_LINKS = {
    "train_val": "https://zenodo.org/records/14963880/files/RRDataset_original_train_val.tar.gz",
    "test": "https://zenodo.org/records/14963880/files/RRDataset_test.tar.gz"
}

def download_and_extract(url, dest_folder):
    filename = url.split('/')[-1]
    filepath = os.path.join(dest_folder, filename)
    extract_marker = filepath.replace(".tar.gz", ".extracted")

    if os.path.exists(filepath):
        print(f"{filename} already exists. Skipping download.")
    else:
        print(f"Downloading {filename}...")
        response = requests.get(url, stream=True)
        total_size = int(response.headers.get('content-length', 0))
        
        with open(filepath, 'wb') as f, tqdm(total=total_size, unit='B', unit_scale=True) as bar:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
                bar.update(len(chunk))

    if os.path.exists(extract_marker):
        print(f"{filename} already extracted, skip.")
    else:
        print(f"Estraggo {filename}...")
        with tarfile.open(filepath, "r:gz") as tar:
            tar.extractall(path=dest_folder)
        open(extract_marker, 'w').close() 
        print("Done")
```

```python
download_and_extract(ZENODO_LINKS["train_val"], DATA_DIR)
#download_and_extract(ZENODO_LINKS["test"], DATA_DIR)
```
